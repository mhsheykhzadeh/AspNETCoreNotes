***ASP.NET Core*** includes a `simple DI container` that all the framework ***libraries*** use to `register` ***themselves*** and ***their dependencies***

We can `replace` buit-in Asp.net core DI with third-parth DI like Autofac, autoregistration and setter injection.

`RequestDelegateFactory` is part of the minimal API framework that’s **responsible** for `invoking` your minimal API `handlers`.

in asp.net core mvc `DefaultControllerActivator` have responsible of ***create new instance*** of `controllers` using `Dependency Injection Container` for ***inject*** all dependencies and then execute target action
