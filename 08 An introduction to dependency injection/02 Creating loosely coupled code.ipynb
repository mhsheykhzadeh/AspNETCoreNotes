<h5>Coupling (اتصال)</h5>
<p><strong>Coupling</strong> به میزان وابستگی بین ماژول‌ها یا اجزای مختلف یک سیستم نرم‌افزاری اشاره دارد. به عبارت دیگر، coupling به چگونگی ارتباط و تعامل بین کلاس‌ها، متدها یا ماژول‌های مختلف می‌پردازد.</p>
<ul><li><p><strong>Tight Coupling (اتصال محکم):</strong> در این حالت، ماژول‌ها یا اجزاء به شدت به یکدیگر وابسته هستند. تغییر در یک ماژول ممکن است نیاز به تغییرات گسترده در سایر ماژول‌ها داشته باشد. این نوع اتصال باعث کاهش انعطاف‌پذیری و افزایش پیچیدگی نگهداری می‌شود.</p><ul><li><strong>مثال:</strong> اگر کلاس A مستقیماً کلاس B را ایجاد کرده و از متدهای آن استفاده کند، این دو کلاس به شدت به هم متصل هستند.</li></ul></li><li><p><strong>Loose Coupling (اتصال سست):</strong> در این حالت، ماژول‌ها یا اجزاء تا حد امکان مستقل از یکدیگر هستند. تغییر در یک ماژول تأثیر کمی بر سایر ماژول‌ها دارد. این نوع اتصال باعث افزایش انعطاف‌پذیری و کاهش پیچیدگی نگهداری می‌شود.</p></li></ul>

<h5>Cohesion (انسجام)</h5>
<p><strong>Cohesion</strong> به میزان همبستگی و وابستگی داخلی وظایف و اجزاء در یک ماژول یا کلاس اشاره دارد. به عبارت دیگر، cohesion به چگونگی تمرکز یک ماژول یا کلاس بر یک وظیفه خاص و متمرکز بودن وظایف آن می‌پردازد.</p>
<ul><li><p><strong>Low Cohesion (انسجام پایین):</strong> در این حالت، وظایف و مسئولیت‌های مختلف و نامرتبط در یک ماژول یا کلاس جمع شده‌اند. این وضعیت باعث پیچیدگی بیشتر و سختی درک و نگهداری کد می‌شود.</p><ul><li><strong>مثال:</strong> کلاسی که همزمان مسئول انجام عملیات ورودی/خروجی، منطق کسب‌وکار و دسترسی به دیتابیس است.</li></ul></li><li><p><strong>High Cohesion (انسجام بالا):</strong> در این حالت، یک ماژول یا کلاس به یک وظیفه خاص و متمرکز محدود می‌شود. این وضعیت باعث کاهش پیچیدگی و افزایش خوانایی و نگهداری کد می‌شود.</p></li></ul>

One problem that remains is that `RegisterUser` is still tied to an `implementation` **rather than** an `abstraction`.

Coding to `abstractions` (often interfaces) is a common design pattern that helps further `reduce the coupling` of a system

This pattern is particularly useful for `making classes testable`

```python
public interface IEmailSender
{
    public void SendEmail(string username);
}

string RegisterUser(string username, IEmailSender emailSender) 
{
    emailSender.SendEmail(username); 
    return $"Email sent to {username}!";
}
```

Now the ***application code*** is `independent` of the ***implementation***.

**how does the application know** to use `EmailSender` in ***production*** instead of `DummyEmailSender`?  
  
The process of ***telling your DI container*** When you need IEmailSender, use EmailSender is called `registration`.

You register services with a DI container so that it knows `which implementation` to use for each ***requested service***.
