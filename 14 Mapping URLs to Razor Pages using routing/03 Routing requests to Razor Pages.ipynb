***Razor Pages*** use `explicit routing` by creating ***route templates*** `based on conventions`.

ASP.NET Core creates a `route template` for ***every Razor Page*** in your app during app startup, when you call `MapRazorPages()` in Program.cs

***By default***, `each Razor Page` creates a `single route template` ***based on its file path***.

The ***exception*** to this rule is for Razor Pages that are called
`Index.cshtml`. ***Index.cshtml*** pages create `two route templates`, one ending with `Index` and the other `without this suffix`.
