### Razor

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn"> 
<p><strong>Razor</strong> یک syntax برای ترکیب کدهای HTML و C# است که در فریم‌ورک ASP.NET، به ویژه در ASP.NET MVC و Razor Pages، برای تولید محتوای سمت سرور (server-side) استفاده می‌شود. این syntax به توسعه‌دهندگان این امکان را می‌دهد که از کدهای HTML و C# به صورت یکپارچه در یک فایل استفاده کنند. در ادامه ویژگی‌ها و کاربردهای Razor را توضیح می‌دهم:</p>
<h3>ویژگی‌های اصلی Razor</h3>
<ol><li><p><strong>سبک و سریع</strong>: Razor به طور خاص برای سادگی و کارایی طراحی شده است. نحوه نوشتار و قالب آن بسیار ساده است و به توسعه‌دهندگان اجازه می‌دهد که بدون نیاز به کدهای اضافه، منطق سمت سرور را به HTML اضافه کنند.</p></li><li><p><strong>مستقل از HTML و C#</strong>: در Razor می‌توانید کدهای HTML و C# را در کنار هم و بدون نیاز به جداکننده‌های خاص بنویسید. برای مثال، <code>@</code> یک نماد ویژه در Razor است که نشان می‌دهد کد بعد از آن یک کد C# است. این امر به راحتی خواندن و نوشتن کد کمک می‌کند.</p></li><li><p><strong>تسهیل در تولید صفحات پویا</strong>: Razor به شما این امکان را می‌دهد که به سادگی مقادیر متغیرها و داده‌های پویا را درون HTML قرار دهید و صفحات وب داینامیک و تعاملی ایجاد کنید.</p></li><li><p><strong>امنیت</strong>: Razor به طور خودکار از درخواست‌های خطرناک و تزریق کدها جلوگیری می‌کند، زیرا داده‌ها را encode می‌کند، که به جلوگیری از حملات XSS کمک می‌کند.</p></li><li><p><strong>مستقل از ASP.NET MVC</strong>: در حالی که Razor در ASP.NET MVC بسیار مورد استفاده است، اما می‌تواند در سایر پروژه‌ها، از جمله Razor Pages و Blazor نیز به کار گرفته شود.</p></li></ol>
<h3>اجزای اصلی Razor</h3>
<ul><li><p><strong>نماها (Views)</strong>: در ASP.NET MVC و Razor Pages، نماها یا همان <strong>Razor Views</strong> فایل‌های <code>.cshtml</code> هستند که با استفاده از Razor و HTML تولید شده‌اند. این فایل‌ها شامل کدهای HTML و C# برای ایجاد و نمایش داده‌ها به کاربران هستند.</p></li><li><p><strong>طرح‌بندی (Layouts)</strong>: طرح‌بندی‌ها فایل‌هایی در Razor هستند که ساختار کلی صفحه، شامل هدر، ناوبری، فوتر و غیره را مشخص می‌کنند و نماها از این طرح‌بندی برای تنظیم محتوای خود استفاده می‌کنند.</p></li><li><p><strong>Partial Views</strong>: <strong>Partial Views</strong> در Razor برای تکه‌های کوچک‌تر UI استفاده می‌شوند که می‌توانند در بخش‌های مختلف صفحات تکرار شوند، مثل فرم‌ها یا عناصر تعاملی.</p></li></ul>

<h3>تفاوت Razor با سایر Syntaxها</h3>
<p>در مقایسه با سایر موتورهای تولید HTML مانند JSP (در Java) یا ERB (در Ruby)، Razor ساده‌تر و روان‌تر است. این syntax بدون نیاز به یادگیری زبان جدیدی به شما امکان ترکیب HTML و کدهای سمت سرور را می‌دهد و در کنار C#، به دلیل ویژگی‌هایی که به زبان اضافه می‌کند، در ساخت صفحات وب موثر و کاربردی است.</p>
</div>

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">  
<h3>مفاهیم و اصطلاحات مختلف مرتبط با Razor Pages در ASP.NET Core</h3>
<ol><li><p><strong>Razor Pages</strong>: این اصطلاح به یک الگوی صفحه‌محور اشاره دارد که در آن مسیریابی، مدل‌بایندینگ و تولید HTML با استفاده از نمای Razor در کنار هم قرار می‌گیرند. Razor Pages کل فریمورک یا ساختار کاری است که صفحه‌ها را به عنوان اجزای اصلی خود در نظر می‌گیرد.</p></li><li><p><strong>Razor Page</strong>: هر <strong>Razor Page</strong> به عنوان یک صفحه یا endpoint منفرد عمل می‌کند. این صفحه معمولاً از دو فایل تشکیل شده است:</p><ul><li>فایل <code>.cshtml</code> که حاوی نمای Razor (کدهای HTML و C# برای نمایش) است.</li><li>فایل <code>.cshtml.cs</code> که شامل کلاس <strong>PageModel</strong> صفحه است و منطق و عملیات اصلی آن را مدیریت می‌کند.</li></ul></li><li><p><strong>PageModel</strong>: <strong>PageModel</strong> کلاس پشتیبان برای هر Razor Page است که مسئولیت دریافت داده‌ها از درخواست‌ها و بایندینگ مدل‌ها را به عهده دارد. در این کلاس، مدل‌های بایندینگ و هندلرهای صفحه (مانند متدهای GET و POST) تعریف می‌شوند.</p></li><li><p><strong>Page Handler</strong>: هر Razor Page به طور پیش‌فرض تنها یک مسیر (Route) را پشتیبانی می‌کند، اما می‌تواند به انواع مختلفی از درخواست‌های HTTP (مانند GET و POST) پاسخ دهد. هر <strong>Page Handler</strong> مسئولیت رسیدگی به یک نوع درخواست HTTP را بر عهده دارد.</p></li><li><p><strong>Razor View</strong>: <strong>Razor views</strong> یا همان <strong>Razor templates</strong> برای تولید HTML استفاده می‌شوند. در مرحله نهایی پردازش Razor Page، این نماها برای تولید خروجی HTML و ارسال آن به کاربر به کار می‌روند.</p></li></ol>
</div>

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn"> 
<p>در ASP.NET Core فایل‌های <strong><code>.cshtml</code></strong> به کلاس‌های C# تبدیل می‌شوند. این تبدیل توسط <strong>Razor Engine</strong> انجام می‌شود و به طور خودکار در زمان کامپایل یا اولین بار اجرا صورت می‌گیرد.</p>
<h3>نحوه تبدیل <code>.cshtml</code> به کد C#</h3>
<p>وقتی شما یک فایل <code>.cshtml</code> را ایجاد می‌کنید، در واقع دارید از Razor Syntax برای ترکیب HTML و C# استفاده می‌کنید. اما برای اینکه این فایل قابل اجرا باشد، نیاز است که به یک کلاس C# تبدیل شود. این فرآیند شامل چند مرحله است:</p>
<ol><li><p><strong>تجزیه فایل <code>.cshtml</code></strong>: ابتدا، Razor Engine محتوای فایل <code>.cshtml</code> را تجزیه می‌کند و HTML و کدهای C# را از هم جدا می‌کند.</p></li><li><p><strong>تبدیل به کلاس C#</strong>: سپس، Razor این محتوا را به یک کلاس C# تبدیل می‌کند که از کلاس <code>RazorPage&lt;T&gt;</code> ارث‌بری می‌کند. این کلاس شامل متدهایی می‌شود که محتوای HTML را رندر کرده و کد C# را اجرا می‌کند.</p></li><li><p><strong>کامپایل به کد میانی (IL)</strong>: در نهایت، این کلاس C# کامپایل شده و به کد میانی (Intermediate Language یا IL) تبدیل می‌شود که توسط .NET Runtime قابل اجرا است.</p></li></ol>

<h3>چه زمانی این تبدیل انجام می‌شود؟</h3>
<p>این تبدیل می‌تواند در دو زمان مختلف اتفاق بیفتد:</p>
<ul><li><p><strong>هنگام کامپایل (Compile-time)</strong>: در صورتی که از <strong>Precompilation</strong> استفاده کنید، فایل‌های <code>.cshtml</code> هنگام ساختن پروژه به کلاس‌های C# و سپس به IL تبدیل می‌شوند. این روش سرعت اجرای برنامه را افزایش می‌دهد، چون فرآیند تبدیل در زمان اجرا انجام نمی‌شود.</p></li><li><p><strong>هنگام اجرا (Run-time)</strong>: اگر Precompilation فعال نباشد، Razor Engine در اولین بار که به فایل <code>.cshtml</code> نیاز دارد، آن را به کلاس C# تبدیل و کامپایل می‌کند. در نتیجه، این فرآیند ممکن است به مقدار کمی زمان بیشتری در اجرای اولیه نیاز داشته باشد.</p></li></ul>
<h3>کجا این کلاس‌های C# ذخیره می‌شوند؟</h3>
<p>این کلاس‌ها در پشت صحنه و به صورت موقت در حافظه نگهداری می‌شوند و به فایل‌هایی که شما می‌بینید اضافه نمی‌شوند. این یعنی توسعه‌دهنده معمولاً نیازی به مشاهده یا مدیریت این کلاس‌های C# ندارد و Razor Engine به طور کامل آن‌ها را مدیریت می‌کند.</p>
