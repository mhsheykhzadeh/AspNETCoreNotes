<h3>Section</h3>
<p><strong>Sections</strong> یا بخش‌ها، قسمت‌های اختیاری و پویا در Layout هستند که Viewها می‌توانند آن‌ها را پر کنند. با استفاده از Sectionها، می‌توانید بخشی از Layout را برای نمایش محتوای خاصی که از Viewهای مختلف تامین می‌شود، پویا کنید.</p>
</div>

### Partial View

<div dir="rtl" style="margin:auto; width:90%; font-family:vazirmatn">
<h3>1. <strong>تعریف Partial View</strong></h3>
<ul><li>معرفی Partial View و اینکه بخشی از یک صفحه وب است که به طور مستقل رندر می‌شود.</li><li>توضیح اینکه Partial View معمولاً برای بخش‌های تکراری یا اجزای مشترک UI مانند فرم‌ها، لیست‌ها، و اطلاعات تکراری استفاده می‌شود.</li></ul>
<h3>2. <strong>مزایای استفاده از Partial View</strong></h3>
<ul><li><strong>کاهش تکرار کد</strong>: اجزای تکراری مانند فرم‌های ورود یا لیست‌های آیتم‌ها را می‌توان یک بار در یک Partial View تعریف کرد و در چندین View استفاده کرد.</li><li><strong>افزایش خوانایی و نگهداری کد</strong>: با تقسیم‌بندی Viewها به بخش‌های کوچکتر و استفاده از Partial Viewها، کد خواناتر و نگهداری آن ساده‌تر می‌شود.</li><li><strong>بارگذاری سریع‌تر</strong>: اگر Partial Viewها به طور جداگانه و با استفاده از تکنیک‌های بارگذاری جزئی بارگذاری شوند، ممکن است سرعت بارگذاری صفحه نیز بهبود یابد.</li></ul>
<h3>3. <strong>نحوه ایجاد Partial View</strong></h3>
<ul><li>ایجاد Partial View به صورت یک فایل <code>.cshtml</code> و قرار دادن آن در پوشه <code>Shared</code> یا هر پوشه دلخواه دیگر.</li><li>نام‌گذاری فایل‌ها با <strong>"_"</strong> برای نشان دادن اینکه یک Partial View است (مثلاً <code>_PartialViewExample.cshtml</code>).</li></ul>
<h3>4. <strong>نحوه استفاده از Partial View در View اصلی</strong></h3>
<ul><li><p>معرفی دو متد اصلی برای استفاده از Partial View:</p><ul><li><strong><code>@Html.Partial()</code></strong>: برای رندر کردن Partial View بدون ارسال مدل.</li><li><strong><code>@Html.RenderPartial()</code></strong>: مشابه <code>@Html.Partial</code> است، اما به‌جای بازگرداندن HTML، آن را مستقیماً در پاسخ خروجی رندر می‌کند (این متد در نسخه‌های جدیدتر ASP.NET Core حذف شده است).</li></ul></li></ul>

<h3>5. <strong>ارسال داده‌ها به Partial View</strong></h3>
<ul><li>نحوه ارسال داده‌ها به Partial View با استفاده از <strong>ViewData</strong> یا ارسال مستقیم مدل.</li><li>استفاده از مدل خاص برای Partial View و ارسال مدل با استفاده از <code>@Html.Partial("_PartialViewExample", model)</code>.</li></ul>
<h3>6. <strong>استفاده از Partial View به‌صورت Async</strong></h3>
<ul><li>معرفی <strong><code>@Html.PartialAsync()</code></strong> و <strong><code>@await Html.PartialAsync()</code></strong> برای رندر غیرهمزمان Partial Viewها که در بارگذاری جزئی صفحات پیچیده و بزرگ کمک می‌کند.</li>
<li dir="ltr"><code>@await Html.PartialAsync("_PartialViewExample")
</code></li>
</ul>
