<p> OpenAPI یک استاندارد زبان‌مستقل برای توصیف API‌های RESTful است. در اصل، OpenAPI به توصیف ساختار یک سند JSON می‌پردازد که در آن URL‌های موجود در برنامه شما، نحوه فراخوانی آن‌ها و انواع داده‌هایی که برمی‌گردانند، توضیح داده می‌شوند.</p>

<p>OpenAPI، که قبلاً به عنوان <strong>Swagger</strong> شناخته می‌شد، ابزاری است که به شما امکان می‌دهد RESTful API‌های خود را به صورت شفاف و قابل فهم توصیف کنید. این توصیف شامل موارد زیر می‌شود:</p>
<ul><li><strong>URL‌های موجود در برنامه</strong>: OpenAPI مشخص می‌کند که چه مسیرهایی (endpoints) در API شما وجود دارند.</li><li><strong>نحوه فراخوانی</strong>: این استاندارد توضیح می‌دهد که چگونه باید درخواست‌ها را به این مسیرها ارسال کرد (مثلاً از طریق روش‌های HTTP مانند GET، POST و غیره).</li><li><strong>انواع داده‌های برگردانده شده</strong>: OpenAPI مشخص می‌کند که چه نوع داده‌ای (مثلاً JSON، XML) از هر مسیر برگردانده می‌شود.</li></ul>

<ol><li><strong>کاوش برنامه با استفاده از Swagger UI</strong>: Swagger UI یک رابط کاربری است که به شما اجازه می‌دهد تا API‌های خود را از طریق یک مرورگر وب به صورت بصری کاوش کنید. شما می‌توانید با استفاده از این رابط کاربری به راحتی درخواست‌ها را به API ارسال کرده و پاسخ‌ها را مشاهده کنید.</li><li><strong>تولید کلاینت‌های قوی‌نوع</strong>: با استفاده از سند OpenAPI می‌توانید کلاینت‌هایی ایجاد کنید که به صورت خودکار با API شما ارتباط برقرار کنند. این کلاینت‌ها بر اساس نوع داده‌های مشخص شده در سند OpenAPI تولید می‌شوند.</li><li><strong>ادغام خودکار با سرویس‌های ثالث</strong>: سند OpenAPI می‌تواند به سرویس‌های خارجی مانند Azure API Management اجازه دهد که به طور خودکار با API شما ادغام شوند.</li></ol>

<h5>مقایسه OpenAPI با WSDL</h5>
<p>اگر با <strong>WSDL</strong> (Web Service Description Language) در سرویس‌های SOAP در ASP.NET قدیمی آشنایی دارید، می‌توانید OpenAPI را معادل آن در دنیای HTTP/REST در نظر بگیرید. همان‌طور که فایل <code>.wsdl</code> سرویس‌های SOAP را توصیف می‌کرد، سند OpenAPI هم API‌های REST شما را توصیف می‌کند.</p>

<h5>استفاده از OpenAPI در ASP.NET Core</h5>
<p>ASP.NET Core به طور پیش‌فرض از OpenAPI پشتیبانی می‌کند، اما برای بهره‌برداری کامل از این قابلیت‌ها، نیاز به استفاده از یک کتابخانه شخص ثالث دارید. دو کتابخانه معروف برای تولید اسناد OpenAPI عبارتند از:</p>
<ul><li><strong>NSwag</strong></li><li><strong>Swashbuckle</strong></li></ul>
</div>

Add the `Swashbuckle.AspNetCore` NuGet package to your project by using
the NuGet Package Manager in Visual Studio, or use the `.NET CLI` by running

`dotnet add package Swashbuckle.AspNetCore`

`Swashbuckle` uses ASP.NET Core ***metadata services*** to retrieve information about `all the endpoints` in your application and to ***generate an OpenAPI document***.

Then this ***document*** is served by `middleware provided by Swashbuckle`

```python
using System.Collections.Concurrent;

WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

builder.Services.AddEndpointsApiExplorer();

builder.Services.AddSwaggerGen();

WebApplication app = builder.Build();

var _fruit = new ConcurrentDictionary<string, Fruit>();

app.UseSwagger();

app.UseSwaggerUI();

app.MapGet("/fruit/{id}", (string id) =>
    _fruit.TryGetValue(id, out var fruit)
        ? TypedResults.Ok(fruit)
        : Results.Problem(statusCode: 404));

app.MapPost("/fruit/{id}", (string id, Fruit fruit) =>
        _fruit.TryAdd(id, fruit)
        ? TypedResults.Created($"/fruit/{id}", fruit)
        : Results.ValidationProblem(new Dictionary<string, string[]>
          {
              { "id", new[] { "A fruit with this id already exists" } }
          }));

app.Run();
record Fruit(string Name, int Stock);
```

you can run application and go to `/swagger/v1/swagger.json` for openAPI document

You can ***change*** some of the **document values**, such as the `title`, by adding
configuration to the `AddSwaggerGen()` method.

```python
builder.Services.AddSwaggerGen(x =>
    x.SwaggerDoc("v1", new OpenApiInfo()
    {
        Title = "Fruitify",
        Description = "An API for interacting with fruit stock",
        Version = "1.0"
    }));
```
