 RESTful API</h3>

<p>استانداردهایی که برای RESTful API وجود دارد، شامل مجموعه‌ای از اصول و بهترین روش‌ها است که به توسعه‌دهندگان کمک می‌کند تا API‌های کارآمد، مقیاس‌پذیر و قابل نگهداری ایجاد کنند. در زیر، به برخی از این استانداردها و اصول اشاره شده است:</p>

<h5>1. <strong>استفاده از HTTP Methods:</strong></h5>
<ul><li><strong>GET</strong>: برای دریافت داده‌ها از سرور.</li><li><strong>POST</strong>: برای ایجاد یک منبع جدید.</li><li><strong>PUT</strong>: برای به‌روزرسانی یک منبع موجود.</li><li><strong>DELETE</strong>: برای حذف یک منبع.</li><li><strong>PATCH</strong>: برای به‌روزرسانی جزئی یک منبع.</li></ul>

<h5>2. <strong>استفاده از URI:</strong></h5>
<ul><li>URI‌ها باید معنی‌دار و توصیفی باشند.</li><li>از نام‌های جمع برای منابع استفاده کنید.</li><li>از حروف کوچک و خط فاصله استفاده کنید. به عنوان مثال:<ul><li>صحیح: <code>/users</code></li><li>نادرست: <code>/Users</code></li></ul></li></ul>

<h5>3. <strong>استفاده از Status Codes:</strong></h5>
<ul><li><strong>200 OK</strong>: درخواست موفقیت‌آمیز بود.</li><li><strong>201 Created</strong>: منبع جدید با موفقیت ایجاد شد.</li><li><strong>204 No Content</strong>: درخواست موفقیت‌آمیز بود ولی داده‌ای برای بازگشت وجود ندارد.</li><li><strong>400 Bad Request</strong>: درخواست نادرست است.</li><li><strong>401 Unauthorized</strong>: احراز هویت انجام نشده است.</li><li><strong>403 Forbidden</strong>: دسترسی به منبع ممنوع است.</li><li><strong>404 Not Found</strong>: منبع یافت نشد.</li><li><strong>500 Internal Server Error</strong>: خطای سرور.</li></ul>

<h5>4. <strong>استفاده از Headerها:</strong></h5>
<ul><li><strong>Content-Type</strong>: نوع داده‌ای که ارسال یا دریافت می‌شود را مشخص کنید (مثلاً <code>application/json</code>).</li><li><strong>Authorization</strong>: برای احراز هویت و مجوز.</li></ul>

<h5>5. <strong>استفاده از Query Parameters:</strong></h5>
<ul><li>برای فیلتر کردن، مرتب‌سازی و صفحه‌بندی داده‌ها از پارامترهای پرس‌وجو استفاده کنید. به عنوان مثال:<ul><li>فیلتر کردن: <code>/users?age=25</code></li><li>مرتب‌سازی: <code>/users?sort=age</code></li><li>صفحه‌بندی: <code>/users?page=2&amp;limit=10</code></li></ul></li></ul>

<h5>6. <strong>مدیریت نسخه‌ها (Versioning):</strong></h5>
<ul><li>نسخه‌بندی API‌ها برای حفظ سازگاری بسیار مهم است. روش‌های مختلفی برای نسخه‌بندی وجود دارد:<ul><li>در URI: <code>/v1/users</code></li><li>در Header: <code>Accept: application/vnd.example.v1+json</code></li></ul></li></ul>

<h5>7. <strong>استفاده از HATEOAS:</strong></h5>
<ul><li>(Hypermedia As The Engine Of Application State) به این معناست که هر پاسخ باید لینک‌هایی به منابع مرتبط دیگر داشته باشد تا کاربران بتوانند به راحتی به آن‌ها دسترسی پیدا کنند.</li></ul>

<h5>8. <strong>مستندسازی:</strong></h5>
<ul><li>مستندسازی جامع و دقیق API‌ها بسیار مهم است. ابزارهایی مانند Swagger و OpenAPI برای این منظور مفید هستند.</li></ul>

<h5>9. <strong>امنیت:</strong></h5>
<ul><li>استفاده از HTTPS برای محافظت از داده‌ها در حین انتقال.</li><li>استفاده از احراز هویت و مجوز مناسب.</li><li>جلوگیری از حملات CSRF و XSS.</li></ul>

#### Mapping verbs to endpoints

```python
WebApplicationBuilder builder = WebApplication.CreateBuilder(args);
WebApplication app = builder.Build();

app.MapGet("/fruit", () => Fruit.All);

var getFruit = (string id) => Fruit.All[id];
app.MapGet("/fruit/{id}", getFruit);

app.MapPost("/fruit/{id}/", Handlers.AddFruit);

Handlers handlers = new();
app.MapPut("/fruit/{id}", handlers.ReplaceFruit);

app.MapDelete("/fruit/{id}", DeleteFruit);

app.Run();

void DeleteFruit(string id)
{
    Fruit.All.Remove(id);
}

internal class Handlers
{
    public void ReplaceFruit(string id, Fruit fruit)
    {
        Fruit.All[id] = fruit;
    }

    public static void AddFruit(string id, Fruit fruit)
    {
        Fruit.All.Add(id, fruit);
    }
}

record Fruit(string Name, int Stock)
{
    public static readonly Dictionary<string, Fruit> All = new();
};
```

These are ways you can `pass handlers` to an `endpoint` by simulating a simple API for interacting with a collection of Fruit items  
  
- A lambda expression, as in the MapGet("/fruit") endpoint
- A Func<T, TResult> variable, as in the MapGet("/fruit/{id}")
endpoint
- A static method, as in the MapPost endpoint
- A method on an instance variable, as in the MapPut endpoint
- A local function, as in the MapDelete endpoint

```python
using System.Text.Json;

WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

WebApplication app = builder.Build();

var _fruits = new Dictionary<string, Fruit>();

app.MapGet("/fruits", () => _fruits);

app.MapGet("/fruits/{id}", async (string id, HttpContext httpcontext) =>
{
    if (_fruits.TryGetValue(id, out var fruit))
    {
        var json = JsonSerializer.Serialize(fruit);
        httpcontext.Response.StatusCode = 200;
        httpcontext.Response.ContentType = "application/json";
        await httpcontext.Response.WriteAsync(json);
    }
    else
    {
        httpcontext.Response.StatusCode = 404;
        await httpcontext.Response.WriteAsync("Fruit not found");
    }
});

app.MapPost("/fruits/{id}", async (string id, Fruit fruit, HttpContext httpcontext) =>
{
    if (_fruits.TryAdd(id, fruit))
    {
        var json = JsonSerializer.Serialize(fruit);
        httpcontext.Response.StatusCode = 201; // Created
        httpcontext.Response.ContentType = "application/json";
        httpcontext.Response.Headers["Location"] = $"/fruits/{id}";
        await httpcontext.Response.WriteAsync(json);
    }
    else
    {
        httpcontext.Response.StatusCode = 400; // Bad Request
        var errorResponse = new Dictionary<string, string[]>
            {
                { "id", new[] { "A fruit with this id already exists" } }
            };
        var errorJson = JsonSerializer.Serialize(errorResponse);
        httpcontext.Response.ContentType = "application/json";
        await httpcontext.Response.WriteAsync(errorJson);
    }
});

app.Run();

record Fruit(string Name, int stock);
```

### Generating responses with IResult

the `endpoint middleware` handles each return type as follows

- `void` or `Task`—The endpoint returns a `200` response with ***no body***.
  
- `string` or `Task<string>`—The endpoint returns a `200` response with the
`string serialized` to the **body** as `text/plain`.  
  
- `IResult` or `Task<IResult>`—The ***endpoint*** executes the `IResult.ExecuteAsync` method. ***Depending on the implementation***, this type can **customize** the response, ***returning any status code***.  
  
- `T` or `Task<T>`—All other types (such as POCO objects) are serialized to
**JSON** and returned in the body of a **200** response as application/json.

```python
WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

WebApplication app = builder.Build();

var _fruits = new Dictionary<string, Fruit>();

app.MapGet("/fruits", () => { });

app.MapGet("/fruits/{id}", (string id, HttpContext httpcontext) =>
{
    _fruits.TryGetValue(id, out var fruit);
    return fruit;
});

app.MapPost("/fruits/{id}", (string id, Fruit fruit, HttpContext httpcontext) =>
{
    _fruits.TryAdd(id, fruit);
    return "fruit created";
});

app.Run();

record Fruit(string Name, int stock);
```

#### Returning status codes with Results and TypedResults

ASP.NET Core exposes the simple **static** helper types `Results` and `TypedResults` in the namespace `Microsoft.AspNetCore.Http`.  
  
You can create a **response** with `common status codes`, optionally including a **JSON body**.  
  
Each of the methods on **Results** and **TypedResults** returns an `implementation of IResult`

***Results*** and **TypedResults** perform the `same function`, The only difference is that the **Results** methods return an `IResult`, whereas ***TypedResults*** return a concrete generic type, such as `Ok<T>`.  
  
the ***generic types*** are easier to use in `unit tests` and in `OpenAPI documentation`

```python
using System.Collections.Concurrent;

WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

WebApplication app = builder.Build();

var _fruit = new ConcurrentDictionary<string, Fruit>();

app.MapGet("/fruits", () => _fruit);

app.MapGet("/fruits/{id}", (string id) =>
    _fruit.TryGetValue(id, out var fruit)
        ? TypedResults.Ok(fruit)
        : Results.NotFound()); //404

app.MapPost("/fruits/{id}", (string id, Fruit fruit) =>
    _fruit.TryAdd(id, fruit)
        ? TypedResults.Created($"/fruits/{id}", fruit) //201
        : Results.BadRequest(new
        { id = "A fruit with this id already exists" })); //400

app.MapPut("/fruits/{id}", (string id, Fruit fruit) =>
{
    _fruit[id] = fruit;
    return Results.NoContent();//204
});

app.MapDelete("/fruits/{id}", (string id) =>
{
    _fruit.TryRemove(id, out _);

    return Results.NoContent();//204
});

app.Run();

record Fruit(string Name, int stock);
```

`Results` and `TypedResults` include methods for **all the common status code**
results you could need.  
  
but if you don’t want to use them for some reason, you can always ***set*** the status code yourself ***directly*** on the `HttpResponse`

```python
using System.Net.Mime;

WebApplicationBuilder builder = WebApplication.CreateBuilder(args);
WebApplication app = builder.Build();

//HttpResponse is one of the special types that minimal APIs know to 
//inject into your endpoint handlers

app.MapGet("/teapot", (HttpResponse response) =>
{
    response.StatusCode = 418;
    response.ContentType = MediaTypeNames.Text.Plain;
    return response.WriteAsync("I'm a teapot!"); 
});
app.Run();
```

#### Returning useful errors with Problem Details

`Problem Details` is a web specification or providing `machine-readable` errors for **HTTP APIs**.

`Results.Problem()` and `Results.ValidationProblem()` methods for generating **Problem Details responses** from minimal APIs

***Problem()*** defaults to a `500` status code, ***ValidationProblem()*** defaults to a `400` status

```python
using System.Collections.Concurrent;

WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

WebApplication app = builder.Build();

var _fruit = new ConcurrentDictionary<string, Fruit>();

app.MapGet("/fruits", () => _fruit);

app.MapGet("/fruits/{id}", (string id) =>
    _fruit.TryGetValue(id, out var fruit)
        ? TypedResults.Ok(fruit)
        : Results.Problem(statusCode: 404));

app.MapPost("/fruits/{id}", (string id, Fruit fruit) =>
    _fruit.TryAdd(id, fruit)
        ? TypedResults.Created($"/fruit/{id}", fruit)
        : Results.ValidationProblem(new Dictionary<string, string[]>
            {
                {"id", new[] {"A fruit with this id already exists"}}
            }));

app.Run();

record Fruit(string Name, int stock);
```

#### Converting all your responses to Problem Details

A ***minimal API application*** could generate an `error response` in several ways:
- Returning an `error status code` from an ***endpoint handler***
- `Throwing an exception` in an ***endpoint handler***, which is caught by the
`ExceptionHandlerMiddleware` or the `DeveloperExceptionPageMiddleware` and converted to an ***error response***.
- The ***middleware pipeline*** returning a `404 response` because a request **isn’t handled** by an endpoint
- A ***middleware component*** in the pipeline `throwing an exception`
- A ***middleware component*** returning an `error response` because a request requires `authentication`, and ***no credentials were provided***

we learned how to ***handle exceptions*** with the `ExceptionHandlerMiddleware`.  
  
`app.UseExceptionHandler("/error")`;  
  
in ***.NET 7***, you can ***configure*** the `ExceptionHandlerMiddleware` (and `DeveloperExceptionPageMiddleware`) to convert an ***exception*** to a
`Problem Details response` ***automatically***.

```python
WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

builder.Services.AddProblemDetails();

WebApplication app = builder.Build();

if (!app.Environment.IsDevelopment())
{
    app.UseExceptionHandler();
}

app.MapGet("/", void () => throw new Exception());

app.Run();
```

***ASP.NET Core*** includes a feature in ***minimal APIs*** for running ***tangential concerns***: `endpoint filters`.

```python
using ASPNETCoreInAction.Services;
using System.Collections.Concurrent;

WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

builder.Services.AddProblemDetails();

WebApplication app = builder.Build();

var _fruit = new ConcurrentDictionary<string, Fruit>();

app.MapGet("/fruits/{id}", (string id) =>
{
   return _fruit.TryGetValue(id, out var fruit)
        ? TypedResults.Ok(fruit)
        : Results.Problem(statusCode: 404);
}).AddEndpointFilter(ValidationHelper.ValidateId); ;

app.Run();

class ValidationHelper
{
    internal static async ValueTask<object?> ValidateId(
        EndpointFilterInvocationContext context, 
        EndpointFilterDelegate next) 
    {
        var id = context.GetArgument<string>(0); // access to parameters 
        if (string.IsNullOrEmpty(id) || !id.StartsWith('f'))
        {
            return Results.ValidationProblem( /*short-circuit*/
            new Dictionary<string, string[]>
            {
                {"id", new[]{"Invalid format. Id must start with 'f'"}}
            });
        }
        return await next(context); 
    }
}
```

#### Adding multiple filters to an endpoint

adding ***multiple filters*** to an endpoint **builds up** a `pipeline`.  
The filters you `add first` are `called first` in the **pipeline**,

```python
using ASPNETCoreInAction.Services;
using System.Collections.Concurrent;

WebApplicationBuilder builder = WebApplication.CreateBuilder(args);

builder.Services.AddProblemDetails();

WebApplication app = builder.Build();

var _fruit = new ConcurrentDictionary<string, Fruit>();

app.MapGet("/fruits/{id}", (string id) =>
{
   return _fruit.TryGetValue(id, out var fruit)
        ? TypedResults.Ok(fruit)
        : Results.Problem(statusCode: 404);
}).AddEndpointFilter(ValidationHelper.ValidateId)
.AddEndpointFilter(async (context, next) =>
{
    app.Logger.LogInformation("Executing filter...");
    object? result = await next(context);
    app.Logger.LogInformation($"Handler result: {result}");
    return result;
});

app.Run();

class ValidationHelper
{
    internal static async ValueTask<object?> ValidateId(
        EndpointFilterInvocationContext context, 
        EndpointFilterDelegate next) 
    {
        var id = context.GetArgument<string>(0); // access to parameters 
        if (string.IsNullOrEmpty(id) || !id.StartsWith('f'))
        {
            return Results.ValidationProblem( /*short-circuit*/
            new Dictionary<string, string[]>
            {
                {"id", new[]{"Invalid format. Id must start with 'f'"}}
            });
        }
        return await next(context); 
    }
}
```

### Implementing the IEndpointFilter interface

```python
WebApplicationBuilder builder = WebApplication.CreateBuilder(args);
WebApplication app = builder.Build();

var _fruit = new ConcurrentDictionary<string, Fruit>();

app.MapGet("/fruit/{id}", (string id) =>
    _fruit.TryGetValue(id, out var fruit)
        ? TypedResults.Ok(fruit)
        : Results.Problem(statusCode: 404))
    .AddEndpointFilter<IdValidationFilter>(); 
    
app.Run();

class IdValidationFilter : IEndpointFilter
{
    public async ValueTask<object?> InvokeAsync(
        EndpointFilterInvocationContext context,
        EndpointFilterDelegate next)
    {
        var id = context.GetArgument<string>(0);
        if (string.IsNullOrEmpty(id) || !id.StartsWith('f'))
        {
            return Results.ValidationProblem(
            new Dictionary<string, string[]>
            {
                {"id", new[]{"Invalid format. Id must start with 'f'"}}
            });
        }
        return await next(context);
    }
}
```

<h5>تشابهات بین فیلترها و میان‌افزارها:</h5>
<ol><li><p><strong>مسیر درخواست‌ها و پاسخ‌ها:</strong> درخواست‌ها در مسیر ورود از طریق یک مؤلفه میان‌افزار عبور می‌کنند و پاسخ‌ها در مسیر خروج دوباره از آن عبور می‌کنند. به طور مشابه، فیلترهای نقطه پایانی می‌توانند کدهایی را قبل از فراخوانی فیلتر بعدی در زنجیره و همچنین بعد از تولید پاسخ اجرا کنند.</p></li><li><p><strong>قطع زنجیره:</strong> میان‌افزارها می‌توانند یک درخواست را با برگرداندن یک پاسخ متوقف کنند، بدون اینکه به میان‌افزارهای بعدی ارسال شود. فیلترها نیز می‌توانند زنجیره فیلترها را با برگرداندن یک پاسخ متوقف کنند.</p></li><li><p><strong>نگرانی‌های متقاطع:</strong> میان‌افزارها معمولاً برای نگرانی‌های متقاطع مانند لاگ‌گیری، پروفایلینگ عملکرد و مدیریت استثناها استفاده می‌شوند. فیلترها نیز برای نگرانی‌های متقاطع مناسب هستند.</p></li></ol>

<h5>تفاوت‌های اصلی بین میان‌افزارها و فیلترها:</h5>
<ol><li><p><strong>اجرای درخواست‌ها:</strong> میان‌افزارها می‌توانند برای تمامی درخواست‌ها اجرا شوند؛ در حالی که فیلترها تنها برای درخواست‌هایی که به "EndpointMiddleware" رسیده و نقطه پایانی مربوطه را اجرا می‌کنند، اجرا می‌شوند.</p></li><li><p><strong>دسترسی به جزئیات بیشتر:</strong> فیلترها به جزئیات بیشتری از نقطه پایانی که اجرا خواهد شد دسترسی دارند، مانند مقدار بازگشتی نقطه پایانی. در حالی که میان‌افزارها به طور کلی این مراحل میانی را نمی‌بینند و فقط پاسخ تولید شده را مشاهده می‌کنند.</p></li><li><p><strong>محدود کردن به زیرمجموعه‌ای از درخواست‌ها:</strong> فیلترها به راحتی می‌توانند به زیرمجموعه‌ای از درخواست‌ها محدود شوند، مانند یک نقطه پایانی خاص یا گروهی از نقاط پایانی. در حالی که میان‌افزارها به طور کلی برای همه درخواست‌ها اعمال می‌شوند (اگرچه می‌توان با استفاده از میان‌افزارهای سفارشی به چیزی مشابه دست یافت).</p></li></ol>

<h5>انتخاب بین میان‌افزار و فیلتر:</h5>
<ul><li><p><strong>عمومیت میان‌افزارها:</strong> میان‌افزارها یک مفهوم عمومی‌تر هستند و بر روی اصول پایه‌تری مانند HttpContext عمل می‌کنند، بنابراین گستره بیشتری دارند. اگر عملکرد مورد نیاز شما هیچ نیازی به جزئیات نقطه پایانی ندارد، باید از یک مؤلفه میان‌افزار استفاده کنید. مدیریت استثناها مثال خوبی است؛ استثناها می‌توانند در هر جایی از برنامه شما رخ دهند و باید آنها را مدیریت کنید، بنابراین استفاده از میان‌افزار مدیریت استثنا منطقی است.</p></li><li><p><strong>جزئیات نقطه پایانی در فیلترها:</strong> از طرف دیگر، اگر به جزئیات نقطه پایانی نیاز دارید یا می‌خواهید برای برخی درخواست‌ها رفتار متفاوتی داشته باشید، باید از یک فیلتر استفاده کنید. اعتبارسنجی مثال خوبی است. همه درخواست‌ها به یک نوع اعتبارسنجی نیاز ندارند. درخواست‌های فایل‌های ثابت، مثلاً، نیاز به اعتبارسنجی پارامترها ندارند، به طوری که درخواست‌های یک نقطه پایانی API دارند.</p></li></ul>
