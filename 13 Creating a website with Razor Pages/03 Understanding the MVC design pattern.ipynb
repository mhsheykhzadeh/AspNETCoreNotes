<p> الگوی طراحی <strong>MVC (Model-View-Controller)</strong>  یکی از الگوهای رایج در طراحی اپلیکیشن‌هایی با رابط کاربری (UI) است. هدف اصلی این الگو جداسازی منطق برنامه، داده‌ها و نمایش آن‌ها در رابط کاربری است.</p>

<ol><li><p><strong>MVC الگوی رایج برای برنامه‌های UI:</strong><br>الگوی MVC به طور گسترده برای طراحی برنامه‌هایی که رابط کاربری دارند، مثل اپلیکیشن‌های وب و دسکتاپ استفاده می‌شود. این الگو به توسعه‌دهندگان کمک می‌کند بخش‌های مختلف یک اپلیکیشن را به خوبی از هم تفکیک کنند.</p></li><li><p><strong>تفسیرهای مختلف MVC:</strong><br>الگوی MVC به شکل‌های مختلفی تعبیر و پیاده‌سازی شده است. در هر پیاده‌سازی، ممکن است تمرکز بر جنبه‌های خاصی از این الگو بیشتر باشد. با این حال، همه نسخه‌های MVC یک مفهوم بنیادی را حفظ می‌کنند: <strong>جداسازی مدیریت داده‌ها از نمایش آن‌ها</strong>.</p></li><li><p><strong>ریشه MVC در اپلیکیشن‌های دسکتاپ (GUI):</strong><br>MVC در ابتدا برای اپلیکیشن‌های دسکتاپ با رابط گرافیکی (GUI) طراحی شده بود، نه اپلیکیشن‌های وب. به همین دلیل، برخی از اصطلاحات و پارادایم‌های استفاده شده در توضیح MVC به محیط‌های GUI مرتبط هستند.</p></li><li><p><strong>هدف اصلی MVC:</strong><br>هدف اصلی این الگو، جدا کردن بخش‌های مربوط به مدیریت و تغییر داده‌ها (یعنی منطق کسب‌وکار) از نمایش و نحوه ارائه آن داده‌ها در رابط کاربری است. این جداسازی باعث می‌شود تا:</p><ul><li><strong>Model:</strong> مسئول مدیریت داده‌ها و منطق کسب‌وکار باشد.</li><li><strong>View:</strong> مسئول نمایش داده‌ها در رابط کاربری باشد.</li><li><strong>Controller:</strong> واسطی بین Model و View باشد که درخواست‌های کاربر را مدیریت می‌کند و با مدل و ویو ارتباط برقرار می‌کند.</li></ul></li></ol>

<ol><li><p><strong>MVC در برنامه‌های وب:</strong></p><ul><li>الگوی <strong>MVC (Model-View-Controller)</strong> به شکل گسترده‌ای در برنامه‌های وب استفاده می‌شود. در این الگو:<ul><li><strong>Model</strong> مسئول مدیریت داده‌ها و منطق کسب‌وکار است.</li><li><strong>View</strong> مسئول نمایش داده‌ها به کاربر است.</li><li><strong>Controller</strong> درخواست‌های کاربر را مدیریت می‌کند و بین Model و View ارتباط برقرار می‌کند.</li></ul></li></ul></li><li><p><strong>ASP.NET Core MVC:</strong></p><ul><li><strong>ASP.NET Core MVC</strong> یک فریم‌ورک در ASP.NET Core است که دقیقاً بر اساس الگوی MVC طراحی شده است.</li><li>در این فریم‌ورک، <strong>کنترلرها</strong> (Controllers) شامل <strong>متدهای اکشن</strong> (Action Methods) هستند که وظیفه دارند درخواست‌های HTTP را دریافت و مدیریت کنند. این متدها، داده‌ها را از Model دریافت کرده و به View ارسال می‌کنند تا به کاربر نمایش داده شوند.</li></ul></li><li><p><strong>Razor Pages:</strong></p><ul><li><strong>Razor Pages</strong> نیز بر اساس الگوی MVC ساخته شده است اما رویکردی ساده‌تر و صفحه‌محور دارد. در Razor Pages به جای استفاده از کنترلرها، از <strong>Page Handlers</strong> (هندلرهای صفحه) برای مدیریت درخواست‌ها استفاده می‌شود.</li><li>با اینکه Razor Pages یک رویکرد متفاوت از ASP.NET Core MVC به نظر می‌رسد، اما در واقع <strong>بر پایه همان فریم‌ورک MVC</strong> بنا شده است. Razor Pages از ASP.NET Core MVC در پس‌زمینه برای مدیریت منطق MVC استفاده می‌کند.</li></ul></li><li><p><strong>تفاوت‌ها:</strong></p><ul><li><strong>ASP.NET Core MVC:</strong> از کنترلرها و متدهای اکشن استفاده می‌کند تا منطق درخواست‌ها را مدیریت کند.</li><li><strong>Razor Pages:</strong> به صورت صفحه‌محور است و به جای کنترلرها از Page Handlers استفاده می‌کند، اما در پشت صحنه همچنان از ASP.NET Core MVC استفاده می‌شود.</li></ul></li></ol>
