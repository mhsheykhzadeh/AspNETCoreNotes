<p>ASP.NET Core MVC و Razor Pages هر دو فریم‌ورک‌هایی برای ساخت وب اپلیکیشن‌ها در ASP.NET Core هستند، اما هر کدام برای موارد استفاده و معماری‌های مختلف طراحی شده‌اند. تفاوت‌های اصلی آن‌ها به شرح زیر است:</p>

<h3>1. <strong>ASP.NET Core MVC:</strong></h3>
<ul><li><strong>الگوی معماری:</strong> مبتنی بر الگوی Model-View-Controller (MVC) است. در این الگو، کنترلرها مسئول مدیریت درخواست‌ها، منطق کسب‌وکار و ارسال داده به ویوها (View) هستند.</li><li><strong>ساختار:</strong> معمولاً ساختاری پیچیده‌تر و ماژولار دارد، زیرا کنترلرها و ویوها در فایل‌های جداگانه سازماندهی شده‌اند.</li><li><strong>استفاده:</strong> برای اپلیکیشن‌هایی که نیاز به معماری لایه‌ای و جداسازی کامل منطق کسب‌وکار از نمایش دارند، مناسب است. به‌ویژه برای پروژه‌های بزرگ‌تر که مدیریت پیچیدگی ضروری است.</li><li><strong>مثال کاربردی:</strong> زمانی که اپلیکیشن شما شامل مجموعه‌ای از صفحات پیچیده یا قابلیت‌هایی مثل API‌ها و منطق کسب‌وکار گسترده است.</li></ul>
<h3>2. <strong>ASP.NET Core Razor Pages:</strong></h3>
<ul><li><strong>الگوی معماری:</strong> Razor Pages یک مدل صفحه‌محور است که تمرکز بیشتری بر روی ساده‌سازی توسعه صفحات دارد. هر صفحه Razor یک PageModel (مشابه کنترلر) دارد که داده‌ها و منطق صفحه را مدیریت می‌کند.</li><li><strong>ساختار:</strong> ساده‌تر است، چرا که هر صفحه شامل کد نمایش (Razor) و PageModel مربوطه است که در همان مکان تعریف می‌شوند یا به صورت جداگانه اما نزدیک به هم قرار دارند.</li><li><strong>استفاده:</strong> برای اپلیکیشن‌های کوچکتر یا فرم‌محور مناسب است که هر صفحه شامل یک منطق ساده و اختصاصی است. Razor Pages به دلیل ساختار ساده‌تر، مناسب‌تر برای پروژه‌هایی است که نیاز به انعطاف‌پذیری زیاد در معماری ندارند.</li><li><strong>مثال کاربردی:</strong> مناسب برای صفحات ساده با فرم‌های ثبت‌نام، ورود، یا فرم‌های تماس.</li></ul>
<h3>مقایسه کلی:</h3>
<ul><li><strong>پیچیدگی:</strong> ASP.NET Core MVC برای پروژه‌های پیچیده‌تر و بزرگ‌تر مناسب‌تر است، در حالی که Razor Pages برای پروژه‌های ساده‌تر یا اپلیکیشن‌های با منطق صفحه‌محور کارآمدتر است.</li><li><strong>سهولت استفاده:</strong> Razor Pages به دلیل یکپارچه‌سازی فایل‌های ویو و منطق، توسعه سریع‌تر و ساده‌تری را ارائه می‌دهد.</li><li><strong>کنترلرها:</strong> در MVC، کنترلرها بخش اصلی مدیریت منطق و درخواست‌ها هستند، در حالی که در Razor Pages، هر صفحه به‌طور مستقل منطق مربوط به خودش را دارد.</li></ul>

### Using the Web Application template

1. Choose Create a New Project from the splash screen or choose `File` >
`New` > `Project` from the main Visual Studio screen.

2.  From the list of templates, choose `ASP.NET Core Web App(Razor Pages)`, ensuring
you select the C# language template.

#### Creating a new Razor Page application with the .NET CLI

- `dotnet new sln -n WebApplication1`  
  
- `dotnet new razor -o WebApplication1`
- `dotnet sln add WebApplication1`

you need ***only add the middleware*** to the pipeline and specify that you want
to ***use Razor Page endpoints*** by calling `MapRazorPages`.

For each request, the ***routing middleware*** uses the ***request’s URL*** to determine `which Razor Page to invoke`.

<h3>1. <strong>مسیر URL و Map شدن درخواست به صفحه:</strong></h3>
<p>در Razor Pages، مسیر هر صفحه به صورت خودکار از مکان فیزیکی فایل Razor (.cshtml) در پروژه تعیین می‌شود. برای مثال:</p>
<ul><li>فرض کنید یک فایل Razor Page با نام <code>Index.cshtml</code> در مسیر <code>Pages/Products</code> قرار دارد.</li><li>در این صورت، درخواست URL با مسیر <code>/Products/Index</code> یا <code>/Products</code> به این صفحه <strong>Map</strong> خواهد شد.</li></ul>
<p>مسیر URL به صورت مستقیم با ساختار فیزیکی فایل‌های Razor Page همخوانی دارد، مگر اینکه به صورت دستی تنظیم شود.</p>
<h3>2. <strong>ساختار یک صفحه Razor Page:</strong></h3>
<p>هر صفحه Razor شامل دو بخش است:</p>
<ul><li><strong>فایل <code>PageName.cshtml</code>:</strong> این فایل شامل کدهای HTML و Razor برای نمایش صفحه است.</li><li><strong>فایل <code>PageName.cshtml.cs</code>:</strong> این فایل که به عنوان مدل صفحه (PageModel) شناخته می‌شود، شامل منطق و هندلینگ درخواست‌های صفحه (GET، POST و ...) است.</li></ul>
<p>به عنوان مثال، برای صفحه‌ای با نام <code>Index</code>، این دو فایل وجود دارند:</p>
<ul><li><code>Pages/Index.cshtml</code></li><li><code>Pages/Index.cshtml.cs</code></li></ul>
<p>در اینجا، درخواست‌ها بر اساس نوع (GET، POST و ...) به متدهای مربوط در PageModel <strong>Map</strong> می‌شوند.</p>
</div>

The `@page` directive on the ***first line of the Razor Page*** is the most important. This directive **must always be placed on the first line** of the file, as it `tells ASP.NET Core` that the `.cshtml` file ***is a Razor Page***. Without it, you won’t be able to view your page correctly.

The next line of the Razor Page defines ***which PageModel*** in your project the
Razor Page ***is associated with***  
  
`@model PrivacyModel`

### Handling request logic with page models and handlers

Razor Pages also have an associated `page model`. By convention, this page model is placed in a file commonly known as a `code-behind` file that has a `.cs extension`

***Page models*** should derive from the `PageModel base class`  
they typically contain*** one or more methods*** called `page handlers` that define ***how to handle requests*** to the Razor Page.

A ***page handler*** is the Razor Pages equivalent of a ***minimal API endpoint handler***;

```python
public class PrivacyModel : PageModel //drived from PageModel
{
    private readonly ILogger<PrivacyModel> _logger;

    public PrivacyModel(ILogger<PrivacyModel> logger)// dependency injection
    {
        _logger = logger;
    }

    public IActionResutlt OnGet()
    {
        return Page()
    } 
}
```

The PrivacyModel contains a ***single handler***, `OnGet`  
  
**Page handlers** are typically named by convention, `based on the HTTP verb`
they respond to.

They return either `void`, indicating that the Razor Page’s template ***should be rendered***, or an `IActionResult` that contains other instructions for generating the response, such as `redirecting the user to a different page`.
