### HTTP (Hypertext Transfer Protocol)

<div dir="rtl" style="width:90%; margin:auto;">
<ul><li><strong>تعریف HTTP</strong>: پروتکلی برای تبادل اطلاعات در وب که بر اساس مدل کلاینت-سرور کار می‌کند.</li><li><strong>تاریخچه</strong>: معرفی شده توسط تیم برنرز لی در سال 1989.</li><li><strong>نسخه‌ها</strong>: HTTP/0.9، HTTP/1.0، HTTP/1.1، HTTP/2، HTTP/3.</li></ul>

</div>

#### Request Line

```python
GET /index.html HTTP/1.1
```

<div dir="rtl" style="width:90%; margin:auto;">
<p>خط درخواست اولین خط از یک پیام HTTP است و باید با یک خط خالی به پایان برسد. اهمیت این خط در این است که سرور براساس اطلاعات موجود در خط درخواست، عملیات مورد نظر را شناسایی و اجرا می‌کند. همچنین، در خطوط درخواست GET و HEAD، اطلاعات پارامترها به صورت پارامترهای جستجو (Query Parameters) در URL قرار می‌گیرند، در حالی که در روش‌های POST و PUT، داده‌ها به طور معمول در بدنۀ درخواست قرار می‌گیرند.</p>

<p>خط درخواست (Request Line) یکی از مهم‌ترین بخش‌های یک درخواست HTTP است که اطلاعات اصلی درباره نوع عملیات مورد نظر کلاینت و منبع هدف را فراهم می‌کند. ساختار خط درخواست شامل سه بخش اصلی است:</p>

<h5>1. روش (Method)</h5>
<p>روش HTTP نشان‌دهنده نوع عملیاتی است که کلاینت قصد دارد بر روی منبع هدف انجام دهد. برخی از روش‌های متداول عبارتند از:</p>
<ul>
<li><code>GET</code></li>
<li><code>POST</code></li>
<li><code>PUT</code></li>
<li><code>DELETE</code></li>
<li><code>HEAD</code></li>
<li><code>OPTIONS</code></li>
<li><code>PATCH</code></li>
</ul>

<h5>2. مسیر (URL)</h5>
<p>مسیر منبع مشخص می‌کند که درخواست به کدام منبع روی سرور اشاره دارد. این مسیر می‌تواند شامل پارامترهای جستجو نیز باشد. به عنوان مثال:</p>
<ul><li><code>/index.html</code></li><li><code>/api/users?id=123</code></li></ul>

<h5>3. نسخه پروتکل (HTTP Version)</h5>
<p>نسخه پروتکل نشان‌دهنده نسخه HTTP مورد استفاده برای این درخواست است. نسخه‌های متداول شامل:</p>
<ul><li><strong>HTTP/1.0</strong></li><li><strong>HTTP/1.1</strong></li><li><strong>HTTP/2</strong></li><li><strong>HTTP/3</strong></li></ul>
</div>

<h3>Methods</h3>
<div dir="rtl" style="width:90%; margin:auto;">
<p>Idempotent (ایدیمپوتنت) یک ویژگی مهم در متدهای HTTP است که بیان می‌کند اجرای چندباره یک درخواست با یک متد مشخص باید همان نتیجه را داشته باشد که اجرای یک‌باره آن دارد. به عبارت دیگر، اگر یک متد ایدیمپوتنت باشد، ارسال مکرر همان درخواست تأثیر اضافی یا تغییری در وضعیت نهایی سرور نخواهد داشت.</p>

<h4>1. GET</h4>
<ul><li><strong>کاربرد</strong>: برای درخواست داده از سرور. این متد برای دریافت منابع مانند صفحات وب، تصاویر، و داده‌های API استفاده می‌شود.</li><li><strong>ویژگی‌ها</strong>:<ul><li><strong>بدنه درخواست ندارد</strong>.</li><li><strong>ایدیمپوتنت</strong> (Idempotent): تکرار چندباره آن نتیجه یکسانی خواهد داشت.</li><li><strong>محفوظ‌سازی نمی‌کند</strong>: فقط داده‌ها را دریافت می‌کند بدون تغییر یا ذخیره‌سازی در سرور.</li></ul></li></ul>

<h4>2. POST</h4>
<ul><li><strong>کاربرد</strong>: برای ارسال داده به سرور به منظور ایجاد یک منبع جدید یا پردازش داده‌ها.</li><li><strong>ویژگی‌ها</strong>:<ul><li><strong>بدنه درخواست دارد</strong>: داده‌ها معمولاً در قالب JSON، XML، یا فرم‌های urlencoded ارسال می‌شوند.</li><li><strong>غیر ایدیمپوتنت</strong>: تکرار چندباره آن ممکن است منجر به ایجاد منابع یا تغییرات متفاوت شود.</li><li><strong>می‌تواند منجر به تغییر وضعیت سرور شود</strong>.</li></ul>
</li></ul>
<h4>3. PUT</h4>
<ul><li><strong>کاربرد</strong>: برای به‌روزرسانی یک منبع موجود یا ایجاد یک منبع جدید در صورتی که وجود نداشته باشد.</li><li><strong>ویژگی‌ها</strong>:<ul><li><strong>بدنه درخواست دارد</strong>.</li><li><strong>ایدیمپوتنت</strong>: تکرار چندباره آن نتیجه یکسانی خواهد داشت.</li><li><strong>منبع هدف به طور کامل جایگزین می‌شود</strong>.</li></ul></li></ul>

<h4>4. DELETE</h4>
<ul><li><strong>کاربرد</strong>: برای حذف یک منبع مشخص از سرور.</li><li><strong>ویژگی‌ها</strong>:<ul><li><strong>بدنه درخواست ندارد</strong> (معمولاً).</li><li><strong>ایدیمپوتنت</strong>: تکرار چندباره آن نتیجه یکسانی خواهد داشت (منبع هدف حذف می‌شود).</li></ul></li></ul>

<h4>5. HEAD</h4>
<ul><li><strong>کاربرد</strong>: مشابه متد GET، اما فقط سرآیندها (Headers) را دریافت می‌کند و بدنه‌ای در پاسخ ندارد.</li><li><strong>ویژگی‌ها</strong>:<ul><li><strong>بدنه درخواست ندارد</strong>.</li><li><strong>ایدیمپوتنت</strong>.</li><li><strong>مفید برای بررسی متادیتای منابع بدون دانلود کامل آن‌ها</strong>.</li></ul></li></ul>

<h4>6. OPTIONS</h4>
<ul><li><strong>کاربرد</strong>: برای درخواست قابلیت‌های ارتباطی سرور و متدهای مجاز برای یک منبع.</li><li><strong>ویژگی‌ها</strong>:<ul><li><strong>بدنه درخواست ندارد</strong> (معمولاً).</li><li><strong>ایدیمپوتنت</strong>.</li><li><strong>پاسخ شامل سرآیند Allow است که متدهای مجاز را فهرست می‌کند</strong>.</li></ul></li></li></ul>

<h4>7. PATCH</h4>
<ul><li><strong>کاربرد</strong>: برای اعمال تغییرات جزئی به یک منبع موجود.</li><li><strong>ویژگی‌ها</strong>:<ul><li><strong>بدنه درخواست دارد</strong>.</li><li><strong>غیر ایدیمپوتنت</strong> (معمولاً)، اما بسته به کاربرد ممکن است ایدیمپوتنت نیز باشد.</li><li><strong>منبع به طور جزئی به‌روزرسانی می‌شود</strong>.</li></ul></li></ul>

### HTTP Headers

<div dir="rtl" style="width:90%; margin:auto;">
<p>سرآیندهای HTTP (HTTP Headers) بخشی مهم از پروتکل HTTP هستند که اطلاعات اضافی در مورد درخواست‌ها و پاسخ‌های HTTP فراهم می‌کنند. این اطلاعات می‌توانند شامل نوع محتوا، طول محتوا، وضعیت احراز هویت، اطلاعات کش (Cache)، و بسیاری جزئیات دیگر باشند. سرآیندهای HTTP به چند دسته تقسیم می‌شوند:</p>

<h4>1. سرآیندهای عمومی (General Headers)</h4>
<p>این سرآیندها برای همه پیام‌های HTTP (درخواست و پاسخ) مشترک هستند و اطلاعات کلی در مورد پیام ارائه می‌دهند.</p>
<ul><li><strong>Cache-Control</strong>: مدیریت کش.<ul><li>مثال: <code>Cache-Control: no-cache</code></li></ul></li><li><strong>Connection</strong>: کنترل اتصال.<ul><li>مثال: <code>Connection: keep-alive</code></li></ul></li><li><strong>Date</strong>: تاریخ و زمان ارسال پیام.<ul><li>مثال: <code>Date: Wed, 21 Oct 2015 07:28:00 GMT</code></li></ul></li></ul>

<h4>2. سرآیندهای درخواست (Request Headers)</h4>
<p>این سرآیندها اطلاعات اضافی خاص یک درخواست HTTP را ارائه می‌دهند.</p>
<ul><li><strong>Accept</strong>: نوع رسانه‌هایی که کلاینت می‌پذیرد.<ul><li>مثال: <code>Accept: text/html</code></li></ul></li><li><strong>Accept-Encoding</strong>: نوع‌های فشرده‌سازی که کلاینت می‌پذیرد.<ul><li>مثال: <code>Accept-Encoding: gzip, deflate</code></li></ul></li><li><strong>Authorization</strong>: اطلاعات احراز هویت.<ul><li>مثال: <code>Authorization: Basic QWxhZGRpbjpvcGVuIHNlc2FtZQ==</code></li></ul></li><li><strong>Host</strong>: نام دامنه سرور.<ul><li>مثال: <code>Host: www.example.com</code></li></ul></li><li><strong>User-Agent</strong>: اطلاعات مربوط به مرورگر یا کلاینت.<ul><li>مثال: <code>User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3</code></li></ul></li></ul>

<h4>3. سرآیندهای پاسخ (Response Headers)</h4>
<p>این سرآیندها اطلاعات اضافی خاص یک پاسخ HTTP را ارائه می‌دهند.
</p>

<ul><li><strong>Access-Control-Allow-Origin</strong>: کنترل دسترسی از منابع مختلف.<ul><li>مثال: <code>Access-Control-Allow-Origin: *</code></li></ul></li><li><strong>Content-Encoding</strong>: نوع فشرده‌سازی محتوا.<ul><li>مثال: <code>Content-Encoding: gzip</code></li></ul></li><li><strong>Content-Type</strong>: نوع رسانه‌ای محتوا.<ul><li>مثال: <code>Content-Type: text/html; charset=UTF-8</code></li></ul></li><li><strong>Set-Cookie</strong>: تنظیم کوکی‌ها.<ul><li>مثال: <code>Set-Cookie: userID=abc123; Path=/; Secure; HttpOnly</code></li></ul></li></ul>

<h4>4. سرآیندهای موجودیت (Entity Headers)</h4>

<p>این سرآیندها اطلاعات در مورد بدنۀ پیام ارائه می‌دهند. این نوع سرآیندها می‌توانند در هر دو درخواست و پاسخ استفاده شوند.</p>
<ul><li><strong>Content-Length</strong>: طول محتوا به بایت.<ul><li>مثال: <code>Content-Length: 348</code></li></ul></li><li><strong>Content-Language</strong>: زبان محتوا.<ul><li>مثال: <code>Content-Language: en-US</code></li></ul></li><li><strong>Content-Location</strong>: مکان منبع مربوط به بدنۀ پیام.<ul><li>مثال: <code>Content-Location: /documents/foo</code></li></ul></li></ul>

<h4>برخی از سرآیندهای مهم و کاربردی</h4>
<h5>سرآیندهای امنیتی</h5>
<ul><li><strong>Strict-Transport-Security (HSTS)</strong>: اجباری کردن استفاده از HTTPS.<ul><li>مثال: <code>Strict-Transport-Security: max-age=31536000; includeSubDomains</code></li></ul></li><li><strong>X-Content-Type-Options</strong>: جلوگیری از MIME type sniffing.<ul><li>مثال: <code>X-Content-Type-Options: nosniff</code></li></ul></li><li><strong>X-Frame-Options</strong>: جلوگیری از نمایش محتوای سایت در فریم‌های دیگر.<ul><li>مثال: <code>X-Frame-Options: DENY</code></li></ul></li></ul>
<h5>سرآیندهای کنترل کش</h5>
<ul><li><strong>Expires</strong>: زمان انقضای محتوا.<ul><li>مثال: <code>Expires: Wed, 21 Oct 2020 07:28:00 GMT</code></li></ul></li><li><strong>ETag</strong>: برچسب نسخه‌ای از منبع.<ul><li>مثال: <code>ETag: "686897696a7c876b7e"</code></li></ul></li><li><strong>Last-Modified</strong>: تاریخ آخرین تغییر محتوا.<ul><li>مثال: <code>Last-Modified: Tue, 15 Nov 1994 12:45:26 GMT</code></li></ul></li></ul>
<h5>سرآیندهای احراز هویت</h5>
<ul><li><strong>WWW-Authenticate</strong>: درخواست اطلاعات احراز هویت از کلاینت.<ul><li>مثال: <code>WWW-Authenticate: Basic realm="Access to the site"</code></li></ul></li><li><strong>Proxy-Authenticate</strong>: درخواست اطلاعات احراز هویت از کلاینت برای استفاده از پروکسی.<ul><li>مثال: <code>Proxy-Authenticate: Basic realm="Access to the proxy"</code></li></ul></li></ul>
</div>

####  مثال درخواست

```python
GET /index.html HTTP/1.1
Host: www.example.com
User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64)
Accept: text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8
Accept-Language: en-US,en;q=0.5
Accept-Encoding: gzip, deflate
Connection: keep-alive
```

```python
HTTP/1.1 200 OK
Date: Wed, 21 Oct 2015 07:28:00 GMT
Content-Type: text/html; charset=UTF-8
Content-Length: 138
Last-Modified: Tue, 15 Nov 1994 12:45:26 GMT
Connection: keep-alive
ETag: "686897696a7c876b7e"
```

### Request Body

<div dir="rtl" style="width:90%; margin:auto;">
<p>بدنه درخواست (Request Body) بخشی از درخواست HTTP است که داده‌های ارسالی از کلاینت به سرور را شامل می‌شود. این بخش معمولاً در متدهای POST، PUT، و PATCH استفاده می‌شود که به داده‌هایی نیاز دارند که باید به سرور ارسال شوند. بدنه درخواست می‌تواند شامل انواع مختلفی از داده‌ها باشد، مانند متن ساده، JSON، XML، فرم‌های urlencoded، و حتی فایل‌های چندبخشی (multipart).</p>

<h4>1. استفاده از بدنه درخواست</h4>

<ul><li>ارسال اطلاعات فرم‌ها.</li><li>ارسال داده‌های JSON برای APIها.</li><li>ارسال فایل‌ها.</li><li>به‌روزرسانی منابع موجود.</li></ul>

<h4>2. انواع فرمت‌های بدنه درخواست</h4>
<h5>2.1. فرم‌های urlencoded</h5>
<p>این فرمت بیشتر برای ارسال داده‌های فرم HTML استفاده می‌شود و داده‌ها به صورت کلید-مقدار با استفاده از &amp; جدا می‌شوند و مقادیر با = از کلیدها جدا می‌شوند.</p>
<code>Content-Type: application/x-www-form-urlencoded</code>
<p>مثلا: </p>
<code>name=John+Doe&age=30&city=New+York</code>

<h5>2.2. JSON</h5>
<p>JSON یکی از رایج‌ترین فرمت‌ها برای ارسال داده‌ها در APIهای مدرن است. داده‌ها به صورت ساختار یافته و به صورت متن ارسال می‌شوند.</p>
<code dir="ltr">Content-Type: application/json</code>
<pre dir="ltr">
{
  "name": "John Doe",
  "age": 30,
  "city": "New York"
}</pre>
<h5>2.3. XML</h5>
<p>XML نیز یک فرمت ساختاریافته برای ارسال داده‌هاست که بیشتر در سرویس‌های وب قدیمی‌تر و برخی از کاربردهای خاص استفاده می‌شود.</p>

<code dir="ltr">Content-Type: application/xml</code>
</div>

```python
<person>
  <name>John Doe</name>
  <age>30</age>
  <city>New York</city>
</person>
```

<div dir="rtl" style="width:90%; margin:auto;">
<h5>2.4. فرم‌های چندبخشی (Multipart Forms)</h5>
<p>این فرمت برای ارسال داده‌هایی که شامل فایل‌ها می‌شوند، استفاده می‌شود. هر بخش از فرم می‌تواند نوع محتوا و داده‌های خاص خود را داشته باشد.</p>
<code>Content-Type: multipart/form-data</code>

</div>

```python
------WebKitFormBoundaryE19zNvXGzXaLvS5C
Content-Disposition: form-data; name="name"

John Doe
------WebKitFormBoundaryE19zNvXGzXaLvS5C
Content-Disposition: form-data; name="file"; filename="example.jpg"
Content-Type: image/jpeg

(binary data)
------WebKitFormBoundaryE19zNvXGzXaLvS5C--
```

<div dir="rtl" style="width:90%; margin:auto;">
<h5>2.5. متن ساده</h5>
<p>این فرمت برای ارسال داده‌های متنی ساده استفاده می‌شود.</p>
<code>Content-Type: text/plain</code>
<pre>This is a plain text message.</pre>

<h5>3. نحوه ارسال بدنه درخواست</h5>

<p>برای ارسال بدنه درخواست، باید نوع محتوا (Content-Type) مشخص شود تا سرور بتواند داده‌ها را به درستی تفسیر کند. این کار با استفاده از سرآیند <code>Content-Type</code> انجام می‌شود.</p>
</div>

```python
POST /api/users HTTP/1.1
Host: www.example.com
Content-Type: application/json
Content-Length: 47

{
  "name": "John Doe",
  "age": 30,
  "city": "New York"
}
```

<div dir="rtl" style="width:90%; margin:auto;">
<h3>4. سرآیندهای مرتبط با بدنه درخواست</h3>
<ul><li><strong>Content-Type</strong>: نوع محتوای بدنه درخواست را مشخص می‌کند.<ul><li>مثال: <code>Content-Type: application/json</code></li></ul></li><li><strong>Content-Length</strong>: طول بدنه درخواست به بایت را مشخص می‌کند.<ul><li>مثال: <code>Content-Length: 47</code></li></ul></li><li><strong>Content-Encoding</strong>: نوع فشرده‌سازی محتوا (اگر وجود داشته باشد) را مشخص می‌کند.<ul><li>مثال: <code>Content-Encoding: gzip</code></li></ul></li><li><strong>Content-Language</strong>: زبان محتوای بدنه درخواست را مشخص می‌کند.<ul><li>مثال: <code>Content-Language: en-US</code></li></ul></li></ul>
</div>

### HTTP Status Codes

<div dir="rtl" style="width:90%; margin:auto;">
<ul><li><strong>1xx - اطلاعاتی (Informational)</strong>: پردازش ادامه دارد.<ul><li>مثال: <code>100 Continue</code></li></ul></li><li><strong>2xx - موفقیت (Success)</strong>: درخواست با موفقیت انجام شد.<ul><li>مثال: <code>200 OK</code></li></ul></li><li><strong>3xx - تغییر مسیر (Redirection)</strong>: نیاز به اقدامات بیشتر برای تکمیل درخواست.<ul><li>مثال: <code>301 Moved Permanently</code></li></ul></li><li><strong>4xx - خطاهای کلاینت (Client Errors)</strong>: خطاهای مربوط به درخواست نامعتبر.<ul><li>مثال: <code>404 Not Found</code></li></ul></li><li><strong>5xx - خطاهای سرور (Server Errors)</strong>: خطاهای داخلی سرور.<ul><li>مثال: <code>500 Internal Server Error</code></li></ul></li></ul>
</div>

### Web Servers

<div dir="rtl" style="width:90%; margin:auto;">
<p>ابزارهایی که در سرورها درخواست‌های HTTP را دریافت کرده و به اپلیکیشن‌های سمت سرور تحویل می‌دهند، به عنوان وب سرور (Web Server) شناخته می‌شوند. وب سرورها نقش حیاتی در زیرساخت وب ایفا می‌کنند و مسئول مدیریت و توزیع درخواست‌های وب هستند. چند نمونه معروف از وب سرورها عبارتند از:</p>
<h4>1. <strong>Apache HTTP Server</strong></h4>
<p>Apache HTTP Server، که اغلب به سادگی به عنوان "Apache" شناخته می‌شود، یکی از محبوب‌ترین و قدیمی‌ترین وب سرورهاست. این سرور متن‌باز است و توسط بنیاد نرم‌افزار Apache توسعه داده می‌شود.</p>

### Kestrel

<div dir="rtl" style="width:90%; margin:auto;">
<p>Kestrel یک وب سرور داخلی (in-process) است که همراه با ASP.NET Core عرضه می‌شود و نقش مهمی در میزبانی و اجرای برنامه‌های ASP.NET Core ایفا می‌کند. در ادامه، به بررسی جزئیات و ویژگی‌های Kestrel پرداخته می‌شود:</p>
<h4>ویژگی‌ها و مزایای Kestrel:</h4>
<ol><li><p><strong>عملکرد بالا</strong>: Kestrel به عنوان یک وب سرور سبک و کارآمد طراحی شده است که قادر به مدیریت تعداد زیادی از درخواست‌ها با عملکرد بالا است. این وب سرور از کتابخانه‌های مدرن و بهینه شده استفاده می‌کند تا کارایی بیشتری داشته باشد.</p></li><li><p><strong>مستقل از پلتفرم</strong>: Kestrel به گونه‌ای طراحی شده است که بر روی تمام پلتفرم‌های اصلی از جمله ویندوز، لینوکس و macOS قابل اجرا باشد. این ویژگی باعث می‌شود که برنامه‌های ASP.NET Core بر روی هر سیستم‌عاملی که دات‌نت پشتیبانی می‌کند، اجرا شوند.</p></li><li><p><strong>پیکربندی ساده</strong>: پیکربندی و استفاده از Kestrel بسیار ساده است. می‌توان آن را به راحتی از طریق کدهای پیکربندی در برنامه ASP.NET Core تنظیم کرد. همچنین قابلیت‌های مختلفی مانند تنظیمات پورت، پروتکل‌ها و محدودیت‌های اتصال را فراهم می‌کند.</p></li><li><p><strong>پشتیبانی از پروتکل‌های مختلف</strong>: Kestrel از پروتکل‌های HTTP/1.1، HTTP/2 و HTTPS پشتیبانی می‌کند. این امکان باعث می‌شود که برنامه‌ها بتوانند با استفاده از آخرین استانداردهای وب به کاربران سرویس دهند.</p></li><li><p><strong>انعطاف‌پذیری</strong>: Kestrel می‌تواند به تنهایی یا در ترکیب با یک سرور وب دیگر مانند IIS یا Nginx به عنوان یک پراکسی معکوس استفاده شود. این انعطاف‌پذیری به توسعه‌دهندگان اجازه می‌دهد تا با توجه به نیازهای خاص خود، بهترین پیکربندی را انتخاب کنند.</p></li></ol>

<h4>سناریوهای استفاده از Kestrel:</h4>
<ol><li><p><strong>استفاده به تنهایی</strong>: Kestrel می‌تواند به تنهایی برای برنامه‌های ساده و کم‌بار استفاده شود. در این حالت، Kestrel به عنوان وب سرور اصلی عمل می‌کند و مستقیماً به درخواست‌های وب پاسخ می‌دهد.</p></li><li><p><strong>استفاده به همراه IIS یا Nginx</strong>: برای برنامه‌های پیچیده‌تر و با بار بالا، Kestrel معمولاً به همراه یک وب سرور پراکسی معکوس مانند IIS یا Nginx استفاده می‌شود. در این حالت، IIS یا Nginx درخواست‌ها را دریافت کرده و آن‌ها را به Kestrel منتقل می‌کنند. این ترکیب به بهبود امنیت، مقیاس‌پذیری و مدیریت بار کمک می‌کند.</p></li></ol>

<h4>محدودیت‌ها و نکات مهم:</h4>

<ol><li><p><strong>امنیت</strong>: Kestrel به تنهایی قابلیت‌های امنیتی پیشرفته‌ای مانند فایروال‌های وب یا احراز هویت پیچیده را ارائه نمی‌دهد. بنابراین، در سناریوهای حساس به امنیت، استفاده از یک پراکسی معکوس توصیه می‌شود.</p></li><li><p><strong>قابلیت‌های مدیریت</strong>: در حالی که Kestrel عملکرد بالایی دارد، اما قابلیت‌های مدیریت سرور مانند ثبت لاگ‌ها، ردیابی و مدیریت فرآیندها به اندازه IIS یا Nginx پیشرفته نیست. بنابراین، در محیط‌های تولیدی با بار بالا، استفاده از یک سرور پراکسی معکوس به همراه Kestrel توصیه می‌شود.</p></li></ol>

### پیکربندی سفارشی Kestrel

```python
var builder = WebApplication.CreateBuilder(args);

// پیکربندی سفارشی Kestrel
builder.WebHost.ConfigureKestrel(options =>
{
    options.ListenAnyIP(5000); // Listen on port 5000
    options.ListenAnyIP(5001, listenOptions =>
    {
        listenOptions.UseHttps(); // Enable HTTPS on port 5001
    });
});

var app = builder.Build();

app.MapGet("/", () => "Hello World!");

app.Run();
```
